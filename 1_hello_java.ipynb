````python
%%plantuml

@startsalt
{
    {T
        +j161(root)
        ++doc
        ++classes
        +++com
        ++++j2
        +++++wk2
        +++++...
        ++src
        +++com
        ++++j2
        +++++wk2
        +++++...
        ++lib
        ++model
    }
}
@endsalt
````

````python
# %load src/com/sd/HelloWorld.java
/*************************************************************************
 * @author jsl
 * @since 20160307
 * to show the use of package
 *
 * root
 *     /src
 *     /classes
 *  javac -d classes -sourcepath src src/com/sd/HelloWorld.java
 *  java -cp .:./classes com/sd/HelloWorld
 *
*************************************************************************/

package com.sd;
public class HelloWorld {
  public void hello() {
    System.out.println("Hello World!");
  }
  public void hello(String name) {
    System.out.printf("Hello %s!", name);
  }
  public static void main(String[] args) {
    System.out.println("Hello, World--");
  }
}
````

````python
!javac -d classes -sourcepath src src/com/sd/HelloWorld.java
!java -cp .:./classes com/sd/HelloWorld
````

````python
/* 
 * @author
 * @since
 * a simple turtle demo
 * taken from http://www.java-online.ch/lego/turtlegrafik.php
 */
package com.sd.turtle;
import ch.aplu.turtle.*;

class TurtleDemoMain {
  Turtle t1 = new Turtle();

  TurtleDemoMain() {
    t1.forward(100);
    t1.right(90);
    t1.forward(100);
    t1.right(90);
    t1.forward(100);
    t1.right(90);
    t1.forward(100);
    t1.right(90);
  }

  public static void main(String[] args) {
    new TurtleDemoMain();
  }
}
````

````python
!javac -cp ./lib/aplu5.jar -d classes/ -sourcepath src/ src/com/sd/turtle/TurtleDemoMain.java 
!java -cp ./classes/:./lib/aplu5.jar com.sd.turtle.TurtleDemoMain
````

### 1.2.3 명령창에서 컴파일 하기


-  컴파일할 때, classpath 명시해주거나 환경변수설정하고 생략

    ```
    $ export CLASSPATH=$CLASSPATH:./classes/
    $ printenv
    $ java com.sd.HelloWorld
    ```
    
-   모든 public class는 하나씩 .java파일에 저장한다. 클래스명과 파일명은
    동일하게. 2개의 클래스를 한 파일에 넣을 경우, 파일명과 동일한
    클래스만 public으로 정의함.

-   javac: java source code (.java) > java byte code (.class)
    $ javac -help
    -verbose\
    -g
    -classpath
    -sourcepath
    -d

-   javac serach order for .class files

    1.  \$JAVA\_HOME/lib

    2.  \$JAVA\_HOME/jre/lib/ext

    3.  \$CLASSPATH

    4.  -classpath

    5.  current directory

-   java debugger(using -g option)

    ```
        $ javac A.java -g
        $ jdb A // stops before entering main()
        > help
        > stop in A.main(java.lang.String[])        //breakpoint
        > run
        > list        //5 lines before and after the current position
        > next
        > step up
        > print obj
        > dump obj
        > watch A.x
        > locals
        > print args[0]
        > stop at A:35
        > cont
    ```
    
-   java: JVM runs java byte code (.class)
